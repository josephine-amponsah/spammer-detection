```python
L = []

with open("../data-new/real_insta_accounts.json", encoding="utf-8") as json_file:
    json_text = json.load(json_file)
    for x in json_text:
        d = {}
        for k, v in x.items():
            if k == 'latestPosts':
                list = []
                for index,y in enumerate(v):
                    list.append(v[index]['likesCount'])
                index +=1
                d["likes_"] = list
            else:
                d[k] = v
        L.append(d)
```

```python
real_extract = pd.DataFrame(L)
```

```python
real_extract["med_likes"] = real_extract["likes_"].apply(np.median)
```

```python
fakes = []

with open("../data-new/fakes_data.json", encoding="utf-8") as json_file:
    json_text = json.load(json_file)
    for x in json_text:
        dict = {}
        for k, v in x.items():
            if k == 'latestPosts':
                list = []
                for index,y in enumerate(v):
                    list.append(v[index]['likesCount'])
                index +=1
                dict["likes_"] = list
            else:
                dict[k] = v
        fakes.append(dict)
```
